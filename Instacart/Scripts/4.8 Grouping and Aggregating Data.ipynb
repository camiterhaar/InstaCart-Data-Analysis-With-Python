```python
# Create loyalty flag
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
# Find stats of prices by loyalty flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'max', 'min']})
```

```python
# Create spending flag
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge[['user_id', 'spending_flag', 'prices']].head(60)
```

```python
# Create order frequency flag
ords_prods_merge['order_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['order_freq'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['order_freq'] <= 20) & (ords_prods_merge['order_freq'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['order_freq'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge[['user_id', 'frequency_flag', 'days_since_prior_order']].head(60)
```

```python
frq_loy= pd.crosstab(ords_prods_merge['frequency_flag'], ords_prods_merge['loyalty_flag'].sort_values(), dropna = False)
```
