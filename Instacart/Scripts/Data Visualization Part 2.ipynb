```python
# Import Libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Create path
path=r'C:\Users\jerem\anaconda\Instacart\Data'
```

```python
# Import data
cust_ords_prods_merge=pd.read_pickle(r'C:\Users\jerem\anaconda\Instacart\Data\cust_ords_prods_merge.pkl')
```

```python
# Create bar chart
cust_ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Create histogram
cust_ords_prods_merge.loc[cust_ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
cust_ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# Create scatterplot
sns.scatterplot(x = 'prices', y = 'prices',data = cust_ords_prods_merge)
```

```python
# Create line chart with sample
np.random.seed(4)
dev = np.random.rand(len(cust_ords_prods_merge)) <= 0.7
```

```python
big = cust_ords_prods_merge[dev]
small = cust_ords_prods_merge[~dev]
```

```python
df_2 = small[['orders_day_of_week','prices']]
```

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Create histogram for time_of_order
hist=cust_ords_prods_merge['time_of_order'].plot.hist(bins = 24)
```

This histogram shows the busiest times of the day. The count of orders between 10am and 5pm (17) are the highest, showing that there are more orders during this time period.

```python
# Create bar chart for loyalty flag
bar=cust_ords_prods_merge['loyalty_flag'].value_counts().sort_index().plot.bar()
```

```python
# Create line chart for prices during times
df_3 = small[['time_of_order','prices']]
```

```python
line_df_3 = sns.lineplot(data = df_3, x = 'time_of_order',y = 'prices')
```

```python
# Create line chart for dependants and age
df_4 = small[['Age','n_dependants']]
```

```python
line_df_4 = sns.lineplot(data = df_4, x = 'Age',y = 'n_dependants')
```

This line chart shows that there is not a distinct pattern/correlation between age and number of dependants

```python
# Create scatterplot for age and spending power
df_5 = small[['Age', 'income']]
```

```python
scatter= sns.scatterplot(x = 'Age', y = 'income',data = df_5)
```

This chart shows that most people make below 200k, but there are several people of all ages who make more than that. It seems that people above 40 have the most spending power
